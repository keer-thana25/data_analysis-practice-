```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Practice-Datasets  (1).csv
```

```python
df=pd.read_csv("Practice-Datasets  (1).csv")
```

```python
df
```

```python
df.head()
df.tail()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['Age'], kde=True, color='skyblue', edgecolor='darkblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(gender_counts, labels=gender_counts.index,
        autopct='%1.1f%%',
        colors=['orange', 'purple'],
        wedgeprops={'edgecolor': 'black'})
plt.title('Percentage Distribution of Gender')
plt.show()
```

```python
city_counts = df['City'].value_counts()

plt.figure(figsize=(5,5))
plt.bar(city_counts.index, city_counts.values,
        color='teal', edgecolor='yellow')
plt.title('Number of Customers in Each City')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x=df['Monthly_Income'], color='violet', linewidth=3)
plt.title('Box Plot of Monthly Income')
plt.xlabel('Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['Spending_Score'], kde=True,
             color='red', edgecolor='white')
plt.title('Distribution of Spending Score')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(df['Purchase_Date'], df['Purchase_Amount'],
         marker='s', color='darkgreen', linewidth=1)
plt.title('Purchase Amount Trend Over Time')
plt.xlabel('Purchase Date')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='Gender', y='Monthly_Income',
            data=df, palette='coolwarm',
            edgecolor='black')
plt.title('Average Monthly Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Monthly Income')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='Product_Category', y='Purchase_Amount',
            data=df, palette='Accent',
            linewidth=1)
plt.title('Box Plot of Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
category_counts = df['Product_Category'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(category_counts, labels=category_counts.index,
        autopct='%1.1f%%',
        colors=['gold', 'cyan', 'magenta', 'lime', 'brown'],
        wedgeprops={'edgecolor': 'black'})
plt.title('Product Category Distribution')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['Purchase_Amount'], bins=20,
             kde=True, color='navy',
             edgecolor='white')
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x='Monthly_Income', y='Spending_Score',
                data=df, hue='Gender',
                palette='Dark2',
                edgecolor='black')
plt.title('Monthly Income vs Spending Score')
plt.xlabel('Monthly Income')
plt.ylabel('Spending Score')
plt.show()
```
